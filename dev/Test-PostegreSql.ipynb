```python
import psycopg2
from psycopg2.extras import RealDictCursor
from retailrag_agent.config.settings import settings
import json

# -------------------------------------------------------------------
# Connexion PostgreSQL
# -------------------------------------------------------------------
conn = psycopg2.connect(
    host=settings.DB_HOST,
    port=settings.DB_PORT,
    dbname=settings.DB_NAME,
    user=settings.DB_USER,
    password=settings.DB_PASSWORD,
)

cursor = conn.cursor(cursor_factory=RealDictCursor)
print("Connexion OK")

# -------------------------------------------------------------------
# 1. RÃ©cupÃ©ration des tables
# -------------------------------------------------------------------
cursor.execute("""
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'public'
  AND table_type = 'BASE TABLE'
ORDER BY table_name;
""")

tables = [row["table_name"] for row in cursor.fetchall()]

schema = {}

# -------------------------------------------------------------------
# 2. Pour chaque table : colonnes + PK + FK
# -------------------------------------------------------------------
for table in tables:
    schema[table] = {
        "columns": [],
        "primary_key": [],
        "foreign_keys": []
    }

    # Colonnes
    cursor.execute("""
    SELECT
        column_name,
        data_type,
        is_nullable
    FROM information_schema.columns
    WHERE table_schema = 'public'
      AND table_name = %s
    ORDER BY ordinal_position;
    """, (table,))
    schema[table]["columns"] = cursor.fetchall()

    # ClÃ© primaire
    cursor.execute("""
    SELECT
        kcu.column_name
    FROM information_schema.table_constraints tc
    JOIN information_schema.key_column_usage kcu
      ON tc.constraint_name = kcu.constraint_name
    WHERE tc.table_schema = 'public'
      AND tc.table_name = %s
      AND tc.constraint_type = 'PRIMARY KEY';
    """, (table,))
    schema[table]["primary_key"] = [row["column_name"] for row in cursor.fetchall()]

    # ClÃ©s Ã©trangÃ¨res
    cursor.execute("""
    SELECT
        kcu.column_name,
        ccu.table_name AS foreign_table,
        ccu.column_name AS foreign_column
    FROM information_schema.table_constraints tc
    JOIN information_schema.key_column_usage kcu
      ON tc.constraint_name = kcu.constraint_name
    JOIN information_schema.constraint_column_usage ccu
      ON ccu.constraint_name = tc.constraint_name
    WHERE tc.table_schema = 'public'
      AND tc.table_name = %s
      AND tc.constraint_type = 'FOREIGN KEY';
    """, (table,))
    schema[table]["foreign_keys"] = cursor.fetchall()

# -------------------------------------------------------------------
# 3. Affichage lisible
# -------------------------------------------------------------------
print("\n================= SCHEMA DATABASE =================\n")

for table, details in schema.items():
    print(f"ðŸ“¦ TABLE: {table}")

    print("  â”œâ”€ Colonnes:")
    for col in details["columns"]:
        nullable = "NULL" if col["is_nullable"] == "YES" else "NOT NULL"
        print(f"  â”‚   - {col['column_name']} ({col['data_type']}, {nullable})")

    print(f"  â”œâ”€ ClÃ© primaire: {details['primary_key']}")

    if details["foreign_keys"]:
        print("  â””â”€ ClÃ©s Ã©trangÃ¨res:")
        for fk in details["foreign_keys"]:
            print(
                f"      - {fk['column_name']} â†’ "
                f"{fk['foreign_table']}.{fk['foreign_column']}"
            )
    else:
        print("  â””â”€ ClÃ©s Ã©trangÃ¨res: Aucune")

    print()

# -------------------------------------------------------------------
# 4. Export JSON (utile pour RAG / cache)
# -------------------------------------------------------------------
with open("db_schema.json", "w", encoding="utf-8") as f:
    json.dump(schema, f, indent=2)

print("SchÃ©ma exportÃ© dans db_schema.json")

# -------------------------------------------------------------------
# Nettoyage
# -------------------------------------------------------------------
cursor.close()
conn.close()
print("Connexion fermÃ©e")
```

```python
import psycopg2
from psycopg2.extras import RealDictCursor
from retailrag_agent.config.settings import settings
import json

# -------------------------------------------------------------------
# Connexion PostgreSQL
# -------------------------------------------------------------------
conn = psycopg2.connect(
    host=settings.DB_HOST,
    port=settings.DB_PORT,
    dbname=settings.DB_NAME,
    user=settings.DB_USER,
    password=settings.DB_PASSWORD,
)

cursor = conn.cursor(cursor_factory=RealDictCursor)
print("Connexion OK")

# -------------------------------------------------------------------
# 1. RÃ©cupÃ©ration des tables ANALYTIQUES UNIQUEMENT
# (exclusion des tables de staging)
# -------------------------------------------------------------------
cursor.execute("""
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'public'
  AND table_type = 'BASE TABLE'
  AND table_name NOT LIKE 'stg_%'
ORDER BY table_name;
""")

tables = [row["table_name"] for row in cursor.fetchall()]

schema = {}

# -------------------------------------------------------------------
# 2. Colonnes / PK / FK
# -------------------------------------------------------------------
for table in tables:
    schema[table] = {
        "columns": [],
        "primary_key": [],
        "foreign_keys": []
    }

    # Colonnes
    cursor.execute("""
    SELECT
        column_name,
        data_type,
        is_nullable
    FROM information_schema.columns
    WHERE table_schema = 'public'
      AND table_name = %s
    ORDER BY ordinal_position;
    """, (table,))
    schema[table]["columns"] = cursor.fetchall()

    # Primary key
    cursor.execute("""
    SELECT
        kcu.column_name
    FROM information_schema.table_constraints tc
    JOIN information_schema.key_column_usage kcu
      ON tc.constraint_name = kcu.constraint_name
    WHERE tc.table_schema = 'public'
      AND tc.table_name = %s
      AND tc.constraint_type = 'PRIMARY KEY';
    """, (table,))
    schema[table]["primary_key"] = [row["column_name"] for row in cursor.fetchall()]

    # Foreign keys
    cursor.execute("""
    SELECT
        kcu.column_name,
        ccu.table_name AS foreign_table,
        ccu.column_name AS foreign_column
    FROM information_schema.table_constraints tc
    JOIN information_schema.key_column_usage kcu
      ON tc.constraint_name = kcu.constraint_name
    JOIN information_schema.constraint_column_usage ccu
      ON ccu.constraint_name = tc.constraint_name
    WHERE tc.table_schema = 'public'
      AND tc.table_name = %s
      AND tc.constraint_type = 'FOREIGN KEY';
    """, (table,))
    schema[table]["foreign_keys"] = cursor.fetchall()

# -------------------------------------------------------------------
# 3. Affichage
# -------------------------------------------------------------------
print("\n================= SCHEMA ANALYTIQUE =================\n")

for table, details in schema.items():
    print(f"ðŸ“¦ TABLE: {table}")
    print("  â”œâ”€ Colonnes:")
    for col in details["columns"]:
        nullable = "NULL" if col["is_nullable"] == "YES" else "NOT NULL"
        print(f"  â”‚   - {col['column_name']} ({col['data_type']}, {nullable})")

    print(f"  â”œâ”€ ClÃ© primaire: {details['primary_key']}")

    if details["foreign_keys"]:
        print("  â””â”€ ClÃ©s Ã©trangÃ¨res:")
        for fk in details["foreign_keys"]:
            print(
                f"      - {fk['column_name']} â†’ "
                f"{fk['foreign_table']}.{fk['foreign_column']}"
            )
    else:
        print("  â””â”€ ClÃ©s Ã©trangÃ¨res: Aucune")

    print()

# -------------------------------------------------------------------
# 4. Export JSON (RAG-ready)
# -------------------------------------------------------------------
with open("db_schema_analytics.json", "w", encoding="utf-8") as f:
    json.dump(schema, f, indent=2)

print("SchÃ©ma analytique exportÃ© dans db_schema_analytics.json")

cursor.close()
conn.close()
print("Connexion fermÃ©e")
```
